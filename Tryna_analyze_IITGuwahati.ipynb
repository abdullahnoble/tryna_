```python
line = df['AssignedLine']
train = df.drop(labels = ['AssignedLine','ID'],axis=1)
train.head()
```

```python
data=pd.concat([train,line],axis=1)
for i in train:
    f,ax=plt.subplots(figsize=(9,7))
    sns.boxplot(x=line,y=i,data=data)
```

```python
data4 = pd.concat([train,line], axis = 'columns')
for col in data4.columns:
    df2 = data4[np.abs(data4[col]-data4[col].mean())<3*data4[col].std()]
df2.shape
        
    
```

```python
line2 = df2['AssignedLine']
train2 = df2.drop(labels = ['AssignedLine'],axis=1)
train2.head()
```

```python
data2=pd.concat([train2,line2],axis=1)
for m in train2:
    f,ax=plt.subplots(figsize=(9,7))
    sns.boxplot(x=line2,y=m,data=data2)
```
